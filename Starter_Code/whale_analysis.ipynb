# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_csv = Path('Resources/whale_returns.csv')
whale_csv_data = pd.read_csv(whale_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
whale_csv_data.head()
```

```python
# Count nulls
whale_csv_data.isna().sum()
```

```python
# Drop nulls
whale_csv_data.dropna(inplace=True)
```

```python
whale_csv_data.head()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_csv = Path('Resources/algo_returns.csv')
algo_csv_data = pd.read_csv(algo_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
algo_csv_data.head()
```

```python
# Count nulls
algo_csv_data.isna().sum()
```

```python
# Drop nulls
algo_csv_data.dropna(inplace=True)
algo_csv_data.head()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_csv = Path('Resources/sp500_history.csv')
sp500_csv_data = pd.read_csv(sp500_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
sp500_csv_data.sort_index(inplace=True)
sp500_csv_data.head()
```

```python
# Check Data Types
sp500_csv_data.dtypes
```

```python
# Fix Data Types
sp500_csv_data['Close'] = sp500_csv_data['Close'].str.replace('$', '')
sp500_csv_data['Close'] = sp500_csv_data['Close'].str.replace(',', '')
# sp500_csv_data["Close"]
sp500_csv_data['Close'] = sp500_csv_data['Close'].astype('float')
```

```python
#sp500_csv_data["Close"] = sp500_csv_data["Close"].astype("float")
```

```python
sp500_csv_data.dtypes
```

```python
# Calculate Daily Returns
sp500_daily_returns = (sp500_csv_data - sp500_csv_data.shift(1)) / sp500_csv_data.shift(1)
sp500_daily_returns.head()
```

```python
# Drop nulls
sp500_daily_returns.dropna(inplace=True)
sp500_csv_data.dropna(inplace=True)
sp500_daily_returns.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_csv_data.rename(columns = {'Close':'sp500_close'}, inplace = True)
sp500_daily_returns.rename(columns = {'Close':'sp500_close'}, inplace = True)
# sp500_csv_data.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_returns = pd.concat([whale_csv_data, algo_csv_data, sp500_daily_returns], axis = "columns", join = "inner")
combined_returns.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
combined_returns_21days = combined_returns.rolling(window=21).std()
# Plot the rolling standard deviation
combined_returns_21days.plot(figsize = (20,10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
combined_correlation = combined_returns.corr()

# Display de correlation matrix
sns.heatmap(combined_correlation, vmin = -1, vmax = 1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
covariance_soros = combined_returns["SOROS FUND MANAGEMENT LLC"].rolling(window = 60).cov(combined_returns["sp500_close"])
                                                                                          
# Calculate variance of S&P 500
variance_sp500 = combined_returns["sp500_close"].rolling(window = 60).var()
                                                                                          
# Computing beta
soros_beta = covariance_soros / variance_sp500
                                                                                          
# Plot beta trend
soros_beta.plot(figsize=(20,10))
                                                                                          
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
rolling_ewm = combined_returns.ewm(halflife = 21).std()
rolling_ewm.plot(figsize = (20,10))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
ann_sharpe_ratios = (combined_returns.mean() * 252) / (combined_returns.std() * np.sqrt(252))
```

```python
# Visualize the sharpe ratios as a bar plot
ann_sharpe_ratios.plot.bar()
```

```python
# Reading data from 2nd stock
costco_csv = Path("Resources/cost_historical.csv")
costco_csv_data = pd.read_csv(costco_csv, index_col='Trade DATE', parse_dates=True, infer_datetime_format=True)
# costco_csv_data.drop(columns = ["Symbol"], inplace = True)
costco_csv_data.head()
```

```python
# Reading data from 3rd stock
google_csv = Path("Resources/goog_historical.csv")
google_csv_data = pd.read_csv(google_csv, index_col='Trade DATE', parse_dates=True, infer_datetime_format=True)
# google_csv_data.drop(columns = ["Symbol"], inplace = True)
google_csv_data.head()
```

```python
# Combine all stocks in a single DataFrame
combined_stocks_returns = pd.concat([apple_csv_data, costco_csv_data, google_csv_data], axis = "rows", join = "inner")
# combined_stocks_returns.columns = ['Apple', "Costco", "Google"]
combined_stocks_returns.head()
```

```python
# Reset Date index
combined_stocks_returns = combined_stocks_returns.reset_index()
combined_stocks_returns.head()
```

```python
# Reorganize portfolio data by having a column per symbol
combined_stocks_returns = combined_stocks_returns.pivot_table(values="NOCP", index="Trade DATE", columns="Symbol")
combined_stocks_returns.head()
```

```python
# Calculate daily returns
stocks_daily_returns = combined_stocks_returns.pct_change()

# Drop NAs
stocks_daily_returns = stocks_daily_returns.dropna().copy()

# Display sample data
stocks_daily_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
portfolio_returns = stocks_daily_returns.dot(weights)
# Display sample data
portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
combined_returns["Daily Returns"] = portfolio_returns
combined_returns.tail()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
combined_returns.dropna(inplace=True)
combined_returns.head()
```
